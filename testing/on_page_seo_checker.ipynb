```python
import advertools as adv
import pandas as pd
 
# The URL you want to extract meta keywords from
url = 'https://techpath.biz/'
 
# Specify the output file. The crawl function saves the crawled data into this file.
output_file = 'output.jl'
 
# Start the crawl (this may take some time depending on the site and amount of URLs to crawl)
adv.crawl(url, output_file=output_file, custom_settings={'CLOSESPIDER_PAGECOUNT': 10})
 
# Load the crawl data into a DataFrame
df = pd.read_json(output_file, lines=True)
 
# Extract meta keywords. Note: The column 'meta_keywords' comes from the crawled data.
# This might need adjustment depending on the actual content of your crawled data.
if 'meta_keywords' in df.columns:
    meta_keywords = df['meta_keywords'].dropna()
 
    # Display meta keywords
    print(meta_keywords)
   
else:
    print("No meta keywords found in the crawled data.")
```

```python
import requests
# import advertools as adv
# import pandas as pd
from bs4 import BeautifulSoup
import re
# import json


class SeoAudit:
    def __init__(self, url):
        self.url = url
        self.response = requests.get(url)
        self.soup = BeautifulSoup(self.response.text, 'html.parser')

    def get_title(self):
        return self.soup.title.text if self.soup.title else None

    def get_first_h1(self):
        h1_tag = self.soup.find('h1')
        return h1_tag.get_text().strip() if h1_tag else None

    def get_meta_description(self):
        meta_desc = self.soup.find('meta', attrs={'name': 'description'})
        return meta_desc['content'] if meta_desc else None
    
    def get_keywords(self):
        keywords = []
        meta_keywords = self.soup.find('meta', attrs={'name': 'keywords'})
        if meta_keywords:
            keywords.extend(meta_keywords['content'].split(','))   
        return keywords

    def find_links(self, link_type):
        if link_type == 'internal':
            pattern = re.compile(r'^https?://(www\.)?techpile.in')
            links = [link['href'] for link in self.soup.find_all('a', href=pattern)]
        elif link_type == 'external':
            pattern = re.compile(r'^(http|www)')
            links = [link['href'] for link in self.soup.find_all('a', href=pattern)]
        return links

    def get_seo_data(self):
        return {
            'title': self.get_title(),
            'metaDescription': self.get_meta_description(),
            'h1': self.get_first_h1(),
            'internalLinks': self.find_links('internal'),
            'internalLinksCount': len(self.find_links('internal')),
            'externalLinks': self.find_links('external'),
            'externalLinksCount': len(self.find_links('external')),
            'keywords': self.get_keywords()
        }

# using the SeoAudit class
url = input('Enter url')
page = SeoAudit(url)
seo_data = page.get_seo_data()
seo_data

# # df = df[['title', 'metaDescription', 'h1', 'internalLinks', 'internalLinksCount', 'externalLinks', 'externalLinksCount', 'keywords']]
# print(df)
# json_object = json.dumps(seo_data)
# print(json_object)
```
