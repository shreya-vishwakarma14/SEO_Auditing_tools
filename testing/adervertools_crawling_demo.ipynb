```python
#Analyzing crawled images from output.jl file 
import advertools as adv
import pandas as pd

crawldf = pd.read_json('output.jl', lines=True) 
img_df = adv.crawlytics.images(crawldf)
img_df.head()
# img_df.notna().mean().sort_values(ascending=False).to_frame().round(2)
```

```python
#Analyzing crawled redirects from output.jl file 
import advertools as adv
import pandas as pd

crawldf = pd.read_json('output.jl', lines=True)
redirect_df = adv.crawlytics.redirects(crawldf)
redirect_df
```
